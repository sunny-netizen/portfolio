```python
def cosine_distance_matrix (array, row_ids):

    # Initialize an array to store the cosine distances
    cosine_distances = np.zeros((array.shape[0], array.shape[0]))

    # Compute the cosine distance between every pair of rows
    for i in range(array.shape[0]):
        for j in range(array.shape[0]):
            cos_sim = np.dot(array[i], array[j]) / (np.linalg.norm(array[i]) * np.linalg.norm(array[j]))
            cos_dist = 1 - cos_sim
            cosine_distances[i, j] = cos_dist
    
    #Add the row IDs to the cosine distances array
    cosine_distances = pd.DataFrame(cosine_distances, index=row_ids, columns=row_ids)

    return cosine_distances

colors_cosine = cosine_distance_matrix(colors_sub, colors_id)
colors_cosine.head()
```
